```python
import pandas as pd
from binance import Client, ThreadedWebsocketManager, ThreadedDepthCacheManager
client = Client(api_key='',
                api_secret='')
```

```python
# pd.DataFrame(L[0])
# L = client.futures_coin_ticker()
# L = client.futures_coin_symbol_ticker()
# L = client.futures_symbol_ticker()
L = client.futures_ticker()
# pd.Series(L[0]).to_frame().T
P = []
for l in L:
    df_pair = pd.Series(l).to_frame().T
    P.append(df_pair)
df = pd.concat(P)
df_p =  df.sort_values('priceChangePercent', ascending=False)
df_p
```
