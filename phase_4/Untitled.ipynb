```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from dateutil.parser import parse
from datetime import timedelta
```

```python
df = pd.read_csv('BTC_02_01_2022_16_58_27_X.txt', sep='|', header=None)
datas = np.array([parse(i) for i in df[0]]) + timedelta(hours=5)
df.index = datas
ds = df[[1, 2, 3, 4]]
ds
```

```python
arr = ds[[2, 4]].mean(axis=1)

plt.figure(figsize=(30, 7))
for x, i in enumerate(arr):
    if i>0.5:
        plt.plot(ds.index[x], i, "o", color='red', markersize=2)
    else:
        plt.plot(ds.index[x], i, "o", color='green', markersize=2)

plt.grid()
```

```python
def f_x(x):
    op1 = 4.5-(9*x)
    op2 = np.e**op1
    op3 = 1+op2
    op4 = 1/op3
    return op4

def f_avrg(x):
    op = 1.58740105196*x-(0.79370052598)
    op2 = op**3
    op3 = op2+0.5
    return op3

y = f_x(arr)
y2 = f_avrg(arr)
```

```python
plt.figure(figsize=(30, 7))
for x, i in enumerate(y):
    if i>0.5:
        plt.plot(ds.index[x], i, "o", color='red', markersize=2)
    else:
        plt.plot(ds.index[x], i, "o", color='green', markersize=2)

plt.grid()
```

```python
plt.figure(figsize=(30, 7))
for x, i in enumerate(y2):
    if i>0.5:
        plt.plot(ds.index[x], i, "o", color='red', markersize=2)
    else:
        plt.plot(ds.index[x], i, "o", color='green', markersize=2)

plt.grid()
```
