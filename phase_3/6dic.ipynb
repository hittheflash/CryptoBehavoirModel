```python
def get_tp_sl_concava(arr, pred, current_price, p_higher=0.195, p_sl=0.25, p_tp=0.48):
    # ls = arr.tolist() # turn into list
    ls = list(arr)
    ls.sort(reverse=True) # sort list
    tp_i = int(round(len(arr)*p_higher, 0)) # higher 19.5%
    top = np.mean(ls[:tp_i]) # mean 19.5%
    # base = np.mean([arr[0], arr[-1]]) # mean bottom
    base = float(current_price)
    unidad = abs((top-base)/base) # normalize arr
    if pred==1:
        sl = base*(1-unidad*p_sl) # SL below bottom
        sl_05 = base*(1+unidad*0.10) # TP over bottom
        tp_15 = base*(1+unidad*0.25) # TP over bottom
        tp = base*(1+unidad*p_tp) # TP over bottom
        return base, tp, tp_15, sl, sl_05, unidad
```
